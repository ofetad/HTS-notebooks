# pathway cne h99

```r
pathid   <- read_tsv(file.path(INFODIR, "pathway_ids.txt"),   col_names = FALSE) 
pathname <- read_tsv(file.path(INFODIR, "pathway_names.txt"), col_names = FALSE) 
pathgene <- read_tsv(file.path(INFODIR, "pathway_genes.txt"), col_names = FALSE)

colnames(pathid) <- "id"
colnames(pathname) <- "name"
colnames(pathgene) <- "genelist"

tmp <- bind_cols(pathid, pathname, pathgene)
tmp <- tmp %>% unite(label, id, name, sep = " | ")

pathname_cne_h99 <- tmp$label
pathgene_cne_h99 <- tmp$genelist

genesets_cne_h99 <- str_split(pathgene_cne_h99, "\\|")
names(genesets_cne_h99) <- pathname_cne_h99
head(genesets_cne_h99, 2)
```

Example:

[CNAG_00057 fructose-1,6-bisphosphatase I Cryptococcus neoformans var. grubii H99](https://www.ncbi.nlm.nih.gov/gene/23883926)

[EC 3.1.3.11 | fructose-bisphosphatase](https://www.genome.jp/dbget-bin/www_bget?ec:3.1.3.11)
